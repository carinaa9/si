```python
import numpy as np
import os
import sys
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname('C:/Users/Asus/si/src/si'), '..')))

from src.si.io.csv import read_csv
from src.si.model_selection.split import train_test_split
from src.si.neighbors.knn_classifier import KNNClassifier
from src.si.linear_model.logistic_regression import LogisticRegression
from src.si.essemble.voting_classifier import VotingClassifier
from src.si.essemble.stacking_classifier import StackingClassifier

from sklearn.preprocessing import StandardScaler
```

Ponto 1

```python
# importar o dataset

dataset = 'C:/Users/Asus/si/datasets/breast-bin.csv'

# ler dataset com read csv

breast = read_csv(dataset, sep=",", label=True, features=True)
breast.print_dataframe()
```

Ponto 2

```python
breast.X = StandardScaler().fit_transform(breast.X)
breast.X
```

Ponto 3

```python
train_breast, test_breast = train_test_split(breast)
print('Treino: ', train_breast.shape())
print('Teste: ', test_breast.shape())
```

```python
train_breast.print_dataframe()
```

```python
test_breast.print_dataframe()
```

Ponto 4

Ponto 7

```python
voting_model.fit(train_breast)
print('Score treino: ', voting_model.score(train_breast))
```

```python
print(voting_model.predict(test_breast))
print()
print('Score teste: ', voting_model.score(test_breast))
```
